```python
#BeautifulSoup -> parts of the page, #requests -> pulls html

import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.indeed.com/jobs?q=business+analyst&l=Los+Angeles%2C+CA')
```

```python
page
```

```python
#to get html -> now the html on that page is assigned to this page

page.text
```

```python
#create an object called soup -> parse (if it was csv, then csv.parser)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
soup
```

```python
#before the dataframe -> create dictionary (data type)
#dictionary -> 'key value pairs' -> unique
#fill the dictionary with the data we scraped

job_data = {
        'title' : [],
        'description': [],
}
```

```python
job_data
```

```python
#find the pattern and 
#find or findAll (expecting multiple)-> will return a list

jobs = soup.findAll('div', attrs={'class' : ['row', 'result', 'clickcard']})
```

```python
#find all -> to loop
#find -> individual attributes

for job in jobs:
   
   title=job.find('a', attrs={'data-tn-element' : 'jobTitle'}).text
   print(title)
```

```python
#description 
#strip = trim (excel)

for job in jobs:
        
    title=job.find('a', attrs={'data-tn-element':'jobTitle'}).text.strip()
    #print(title)
    job_data['title'].append(title)            
    description = job.find('span', attrs={'class':'summary'}).text.strip()
    
    print(description)    
    job_data['description'].append(description)
```

```python
df = pd.DataFrame(job_data)
```
